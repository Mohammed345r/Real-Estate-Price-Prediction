```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Suburb'] = le.fit_transform(data['Suburb'])
data['Type'] = le.fit_transform(data['Type'])
data['Method'] = le.fit_transform(data['Method'])
data['SellerG'] = le.fit_transform(data['SellerG'])
data['CouncilArea'] = le.fit_transform(data['CouncilArea'])
data['Regionname'] = le.fit_transform(data['Regionname'])
data['Propertycount'] = le.fit_transform(data['Propertycount'])
data['CouncilArea'] = le.fit_transform(data['CouncilArea'])
```

```python
nulls = data.isnull().sum()
print(nulls)
```

```python
mean_price = data['Price'].mean()
mean_bedroom = data['Bedroom2'].median()
mean_bathroom = data['Bathroom'].mean()
mean_car = data['Car'].median()
mean_landsize = data['Landsize'].mean()
mean_building_area = data['BuildingArea'].median()
mean_propertycount = data['Propertycount'].mean()
mean_yearbuilt = data['YearBuilt'].mean()
mean_suburb = data['Suburb'].mean()
mean_rooms = data['Rooms'].mean()
mean_method = data['Method'].mean()
mean_councilarea = data['CouncilArea'].mean()


data['Price'].fillna(mean_price, inplace=True)
data['Bedroom2'].fillna(mean_bedroom, inplace=True)
data['Bathroom'].fillna(mean_bathroom, inplace=True)
data['Car'].fillna(mean_car, inplace=True)
data['Landsize'].fillna(mean_landsize, inplace=True)
data['BuildingArea'].fillna(mean_building_area, inplace=True)
data['Propertycount'].fillna(mean_propertycount, inplace=True)
data['YearBuilt'].fillna(mean_yearbuilt, inplace=True)
data['Suburb'].fillna(mean_suburb, inplace=True)
data['Rooms'].fillna(mean_rooms, inplace=True)
data['Method'].fillna(mean_method, inplace=True)
data['CouncilArea'].fillna(mean_councilarea, inplace=True)

```

```python
nulls = data.isnull().sum()
print(nulls)
```
